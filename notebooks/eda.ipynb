```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
conn = sqlite3.connect("../employee_attrition.db")

df = pd.read_sql("SELECT * FROM employees", conn)

df.head()
```

## Attrition Distribution

```python
plt.figure(figsize=(6,6))

df["Attrition"].value_counts().plot.pie(
    autopct="%1.1f%%",
    colors=["#2d6a9f","#ff4444"]
)

plt.title("Attrition Distribution")
plt.ylabel("")
plt.show()
```

```python
## Attrition by Department
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="Department",
    hue="Attrition"
)

plt.title("Attrition by Department")
plt.show()
```

## Attrition vs overtime

```python
plt.figure(figsize=(6,5))

sns.countplot(
    data=df,
    x="OverTime",
    hue="Attrition"
)

plt.title("Attrition vs OverTime")
plt.show()
```

## income vs attrition

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="Attrition",
    y="MonthlyIncome"
)

plt.title("Monthly Income vs Attrition")
plt.show()
```

## Age Distribution

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df["Age"],
    bins=20,
    kde=True
)

plt.title("Age Distribution")
plt.show()
```

## correlation Heatmap

```python
numeric_df = df.select_dtypes(include="number")

plt.figure(figsize=(12,8))

sns.heatmap(
    numeric_df.corr(),
    cmap="coolwarm"
)

plt.title("Feature Correlation")
plt.show()
```

# Final Insights from Exploratory Data Analysis

### 1. Attrition Distribution
The attrition distribution shows that the majority of employees stay with the organization. Approximately *84% of employees remain, while about **16% leave the company*. Although most employees stay, the attrition rate still indicates a noticeable level of employee turnover that organizations should monitor.


### 2. Attrition by Department
The *Sales department records the highest number of employees leaving the company* compared to other departments. In contrast, the *Human Resources department has the lowest attrition levels*. This suggests that factors such as workload, job pressure, or sales targets may contribute to higher attrition in the Sales department.


### 3. Attrition vs Overtime
Employees who *work overtime show a higher likelihood of leaving the company* compared to those who do not work overtime. This indicates that *long working hours and work-life balance issues may be significant contributors to employee attrition*.


### 4. Income vs Attrition
Employees who leave the organization generally have *lower monthly income compared to employees who stay*. This suggests that compensation may play an important role in employee retention, and competitive salary structures could help reduce attrition.

---

### 5. Age Distribution
The age distribution indicates that most employees fall within the *25–40 age range, representing the majority of the workforce. Attrition appears to occur more frequently among **younger employees*, which may be due to career mobility, job switching, or pursuit of better opportunities.


### 6. Correlation Heatmap
The correlation heatmap highlights relationships between numerical features. Variables such as *TotalWorkingYears, JobLevel, and MonthlyIncome show positive correlation, indicating that salary and job level tend to increase with experience. However, no single feature alone strongly determines attrition, suggesting that **employee turnover is influenced by multiple factors working together*.


# Conclusion
The exploratory analysis indicates that *overtime workload, department roles, salary levels, and job satisfaction factors* contribute to employee attrition. Organizations can reduce attrition by improving work-life balance policies, offering competitive compensation, and ensuring employee engagement across departments.

## Project Summary

This exploratory data analysis identified several factors associated with employee attrition. The analysis shows that employees working overtime, employees in the Sales department, and those with lower monthly income tend to have higher attrition rates.

Additionally, younger employees show slightly higher turnover trends, while salary, experience, and job level demonstrate strong correlations. These findings suggest that work-life balance, compensation, and departmental workload may significantly influence employee retention.

The insights from this analysis can help organizations design targeted strategies to improve employee satisfaction and reduce attrition.
